Función para evaluar los modelos

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
import matplotlib.pyplot as plt
import pandas as pd

def evaluate(model, X_train, X_test, y_train, y_test):
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)

    print("Resultados del Entrenamiento: \n===============================")
    clf_report = pd.DataFrame(classification_report(y_train, y_train_pred, output_dict=True))
    print(f"Matrix de Confusión:\n{confusion_matrix(y_train, y_train_pred)}")
    print(f"ACCURACY SCORE:\n{accuracy_score(y_train, y_train_pred):.4f}")
    print(f"Reporte de Clasificación:\n{clf_report}")

    print("Resultados de las Pruebas: \n===============================")
    clf_report = pd.DataFrame(classification_report(y_test, y_test_pred, output_dict=True))
    print(f"Matrix de Confusión:\n{confusion_matrix(y_test, y_test_pred)}")
    print(f"ACCURACY SCORE:\n{accuracy_score(y_test, y_test_pred):.4f}")
    print(f"Reporte de Clasificación:\n{clf_report}")
```

## RANDOM FOREST

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
evaluate(rf, X_train, X_test, y_train, y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.tree import export_graphviz
import graphviz
from IPython.display import display


print("\n=========RANDOM FOREST============")
criterion = ['gini', 'entropy']

params_grid = {
    'n_estimators': [5, 25, 50],
    'max_features': ['auto', 'sqrt'],
    'max_depth': [2, 5, 10, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'bootstrap': [True, False]
}

rf_clf = RandomForestClassifier(random_state=1) #42

rf_cv = GridSearchCV(rf_clf, params_grid, scoring="f1", cv=3, verbose=2, n_jobs=-1)

rf_cv.fit(X_train, y_train)
best_params = rf_cv.best_params_
print(f"Best parameters: {best_params}")

rf_clf = RandomForestClassifier(**best_params)
rf_clf.fit(X_train, y_train)

evaluate(rf_clf, X_train, X_test, y_train, y_test)


importancia_predictores = pd.DataFrame(
                            {'predictor': X_train.columns,
                             'importancia': rf_clf.feature_importances_}
                            )
print("Importancia de los predictores en el modelo")
print("-------------------------------------------")
importancia_predictores.sort_values('importancia', ascending=False)

```

## XGBOOST

```python
data = pd.read_excel("BASE_MODELO_v2.xlsx")
data["Cumplimiento_EESS"] = data["Cumplimiento_EESS"].astype("category")
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.pipeline import Pipeline

X = data.drop('Cumplimiento_EESS', axis=1)
y = data['Cumplimiento_EESS']

print(f'X shape: {X.shape}')
print(f'y shape: {y.shape}')

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state= 12)

pipe = Pipeline([
    ('min_max_scaler', MinMaxScaler()),
    ('std_scaler', StandardScaler())
])
```

```python
from xgboost import XGBClassifier

rf = XGBClassifier()
rf.fit(X_train, y_train)
evaluate(rf, X_train, X_test, y_train, y_test)


from sklearn.model_selection import cross_val_score

# Aplicar validación cruzada
scores = cross_val_score(rf, X_train, y_train, cv=5)
```

```python
from xgboost import XGBClassifier
from sklearn.model_selection import GridSearchCV
from IPython.display import display
import graphviz
# Definir el conjunto de parámetros a buscar
param_grid = {
    'n_estimators': [10, 50],
    'max_depth': [2, None],
    'learning_rate': [0.01, 0.1, 0.2],
    'subsample': [0.9],
    'colsample_bytree': [0.8, 0.9, 1.0]
}


clf = XGBClassifier(random_state=1) #42
grid_search = GridSearchCV(clf, param_grid, scoring='f1', cv=3, verbose=2, n_jobs=-1,error_score='raise')
grid_search.fit(X_train, y_train)

best_params = grid_search.best_params_
clf = XGBClassifier(**best_params, random_state=1) #1
clf.fit(X_train, y_train)


print("\n=========XGBOOST============")
print(f"Best parameters: {best_params}")
evaluate(clf, X_train, X_test, y_train, y_test)


importancia_predictores = pd.DataFrame(
                            {'predictor': X_train.columns,
                             'importancia': clf.feature_importances_}
                            )
print("Importancia de los predictores en el modelo")
print("-------------------------------------------")
importancia_predictores.sort_values('importancia', ascending=False)
```
